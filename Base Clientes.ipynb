```python
import pandas as pd
import numpy as np
import os
from openpyxl import load_workbook
import xlsxwriter
from shutil import copyfile
import time
```

```python
file=input('Ruta del archivo: ')
```

```python
#importamos el arcivho TdF
base=pd.read_excel(file)
```

```python
#ver la lista de clientes
base.Cliente.unique()
```

```python
#Filtramos por el cliente que hayamos elegido de la lista de arriba
base=base.loc[base['Cliente']=='Banco de la Pcia de Córdoba']
base
```

```python
#Filtramos y sacamos los bloqueados
base=base.loc[base['Bloqueado']!='X']
```

```python
#Renombramos las columnas
base.columns = ['AP','Pos','Servicio','Cliente','Dia RO','Pto Planif RO','Recorrido RO','Horario RO','Dia ED','Pto Planif ED','Recorrido ED','Horario ED','Direccion O','Dia RD','Pto Planif RD','Recorrido RD','Horario RD','Dia EO','Pto Planif EO','Recorrido EO','Horario EO', 'Direccion D', 'Bloqueado']
base
```

```python
base=base.reset_index()
```

```python
base=base.drop('index',axis=1)
base
```

```python
#en aquellos casos en que el horario sea 00:00:00 y esté vacío en las celdas de dicha parada se procede a cambiar el horario por NaN
for i in range(0,len(base)):
    if np.isnan(base['Dia RD'][i]) == True:
        base['Horario RD'][i]=np.nan
    if np.isnan(base['Dia EO'][i]) == True:
        base['Horario EO'][i]=np.nan
base
```

```python
#duplicamos las columnas día, nos sirve para la pivot
base['Dia RO2']=base['Dia RO']
base['Dia ED2']=base['Dia ED']
base['Dia RD2']=base['Dia RD']
base['Dia EO2']=base['Dia EO']
```

```python
#lista de columnas que estarán en la tabla pivot
col_ro=['AP','Pos', 'Servicio','Pto Planif RO','Direccion O','Pto Planif ED','Direccion D','Horario RO']
col_ed=['Servicio','Horario ED']
col_rd=['Servicio','Horario RD']
col_eo=['Servicio','Horario EO']
```

```python
#creamos 4 tablas pivot, una por cada parada del servicio y reseteamos el índice
p_ro=pd.pivot_table(base, values = 'Dia RO', columns = 'Dia RO2', index = col_ro, fill_value='', aggfunc = 'sum')
p_ed=pd.pivot_table(base, values = 'Dia ED', columns = 'Dia ED2', index = col_ed, fill_value='', aggfunc = 'sum')
p_rd=pd.pivot_table(base, values = 'Dia RD', columns = 'Dia RD2', index = col_rd, fill_value='', aggfunc = 'sum')
p_eo=pd.pivot_table(base, values = 'Dia EO', columns = 'Dia EO2', index = col_eo, fill_value='', aggfunc = 'sum')
p_ro=p_ro.reset_index()
p_ed=p_ed.reset_index()
p_rd=p_rd.reset_index()
p_eo=p_eo.reset_index()
```

```python
#concatenamos las columnas de días en una misma columna
p_ro['Dias RO']=p_ro.iloc[:,8:].apply(lambda x: ';'.join(x.dropna().astype(str)), axis=1)
p_rd['Dias RD']=p_rd.iloc[:,2:].apply(lambda x: ';'.join(x.dropna().astype(str)), axis=1)
p_ed['Dias ED']=p_ed.iloc[:,2:].apply(lambda x: ';'.join(x.dropna().astype(str)), axis=1)
p_eo['Dias EO']=p_eo.iloc[:,2:].apply(lambda x: ';'.join(x.dropna().astype(str)), axis=1)
```

```python
#dropeamos las columnas con días
p_ro=p_ro[['AP','Pos','Servicio', 'Pto Planif RO','Direccion O','Pto Planif ED',
         'Direccion D','Horario RO','Dias RO']]
p_ed=p_ed[['Servicio','Horario ED','Dias ED']]
p_rd=p_rd[['Servicio','Horario RD','Dias RD']]
p_eo=p_eo[['Servicio','Horario EO','Dias EO']]
```

```python
#unimos los dataframe en uno mismo
from functools import reduce
pivotes=[p_ro,p_ed,p_rd,p_eo]
merge = reduce(lambda  left,right: pd.merge(left,right,on=['Servicio'],
                                            how='outer'), pivotes).fillna('NaN')
```

```python
#reordenamos las columnas
merge = merge[['AP', 'Pos', 'Servicio', 'Pto Planif RO', 'Direccion O', 'Pto Planif ED', 'Direccion D', 'Horario RO','Dias RO','Horario ED','Dias ED','Horario RD','Dias RD','Horario EO','Dias EO']]
merge
```

```python
#Calculamos la frecuencia de cada servicio contando la cantidad de caracteres
merge['Frecuencia']=''
for i in range(0,len(merge)):
    lo=len(merge['Dias RO'][i])
    ld=len(merge['Dias EO'][i])
    lt=lo+ld
    if lt==11:                    #ida 1 vez por semana
        merge['Frecuencia'][i]=5
    elif lt==16 or lt==15:
        merge['Frecuencia'][i]=10
    elif lt==14:                  #ida 2 veces por semana
        merge['Frecuencia'][i]=9
    elif lt==22 or lt==21:
        merge['Frecuencia'][i]=18
    elif lt==17:                  #ida 3 veces por semana
        merge['Frecuencia'][i]=13
    elif lt==28 or lt==27:
        merge['Frecuencia'][i]=26
    elif lt==20:                  #ida 4 veces por semana
        merge['Frecuencia'][i]=18
    elif lt==34:
        merge['Frecuencia'][i]=36
    elif lt==23:                  #ida 5 veces por semana
        merge['Frecuencia'][i]=22
    elif lt==40 or lt==39:
        merge['Frecuencia'][i]=44
```

```python
#"convertimos" de str a int sacando los .0
merge['Dias RO'] = merge['Dias RO'].str.replace('.0','')
merge['Dias ED'] = merge['Dias ED'].str.replace('.0','')
merge['Dias RD'] = merge['Dias RD'].str.replace('.0','')
merge['Dias EO'] = merge['Dias EO'].str.replace('.0','')
```

```python
#sacamos los doble punto y coma, se repite 2 veces para asegurar
merge['Dias RO'] = merge['Dias RO'].str.replace(';;',';')
merge['Dias ED'] = merge['Dias ED'].str.replace(';;',';')
merge['Dias RD'] = merge['Dias RD'].str.replace(';;',';')
merge['Dias EO'] = merge['Dias EO'].str.replace(';;',';')
```
